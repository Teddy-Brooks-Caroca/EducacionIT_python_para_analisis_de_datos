### El día con mayor cantidad de delitos de todo el 2019 (usando groupby).

```python
delitos_caba.groupby("fecha")["cantidad_registrada"].sum().idxmax()
```

### Usar un gráfico de barras para mostrar el top 5 de barrios más peligrosos.

```python
delitos_caba["barrio"].value_counts().head(5)
```

```python
delitos_caba["barrio"].value_counts().head(5).plot.bar()
```

### Usar un gráfico de barras para mostrar el top 5 de barrios más seguros.

```python
delitos_caba["barrio"].value_counts().tail(5)
```

```python
delitos_caba["barrio"].value_counts().tail(5).plot.bar()
```

### Crear una nueva columna llamada "mes", y basarse en ella para averiguar cuál es el mes más inseguro.

```python
delitos_caba["mes"] = pd.to_datetime(delitos_caba["fecha"]).dt.month

delitos_caba.groupby("mes")["cantidad_registrada"].sum().idxmax() 
```

### Para finalizar hacer un gráfico de torta de "tipo de delito", ¿cuál es el que más frecuente según el gráfico?

```python
delitos_caba["tipo_delito"].value_counts().plot.pie()
```
